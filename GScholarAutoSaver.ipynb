```python
import cv2
import numpy as np
import pyautogui
import random
import pyautogui
import time

def find_image_on_screen(target_image_path):
    # 将屏幕截图转换为OpenCV格式
    screenshot = pyautogui.screenshot()
    screenshot_cv = cv2.cvtColor(np.array(screenshot), cv2.COLOR_RGB2BGR)

    # 读取目标图像
    target_image = cv2.imread(target_image_path, cv2.IMREAD_UNCHANGED)

    # 使用模板匹配寻找目标图像
    result = cv2.matchTemplate(screenshot_cv, target_image, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # 设定阈值
    threshold = 0.8
    if max_val < threshold:
        return None  # 如果匹配度低于阈值，则认为没有找到匹配，返回None

    # 计算目标图像的中心位置
    target_height, target_width = target_image.shape[:2]
    center_x = max_loc[0] + target_width // 2
    center_y = max_loc[1] + target_height // 2

    return center_x, center_y

# 使用示例
target_image_path = 'star.JPG'
next_image_path = 'Next.JPG'
n_scroll = 0
while True:
    position = find_image_on_screen(target_image_path)


    if position:
        n_scroll = 0
        print(f"Found image at position: {position}")
        pyautogui.click(position)
        time.sleep(random.randint(1,2))  # 等待弹窗，可能需要调整
        pyautogui.press('esc')  # 关闭弹窗
        time.sleep(random.randint(1,2)) 
        # 可以在这里执行点击操作
    else:
        print(f"Image not found.{n_scroll}")
        pyautogui.scroll(-1000)  # 向下滚动
        time.sleep(random.randint(1,2)) 
        n_scroll += 1
        if n_scroll > 2:
            next_position = find_image_on_screen(next_image_path)
            if next_position:
                pyautogui.click(next_position)
                time.sleep(random.randint(2,4))

```
